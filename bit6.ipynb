```python
df3 = df.merge(df2, on = 'Entry', how = 'left')
```

```python
df3
```

```python
df3['40%'] = [int(i * 0.4) for i in df3.Length]
```

```python
df3
```

```python
N = []
for a, b, c, d in zip(df3.Entry, df3.fin, df3.tipo, df3['40%']):
    if b <= d:
        if c == 'Domain':
            print(a, b, c, d)
            N.append(a)
```
