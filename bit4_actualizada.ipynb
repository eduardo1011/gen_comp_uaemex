## Visualización de resultados

#### Cuando veamos una estructura delimitada por corchetes es una lista  
#### Cuando veamos una estructura delimitada por comillas es una cadena de caracteres  
#### Cuando veamos una estructura delimitada por llaves es un diccionario  
##### Otras estructuras de datos son los: dataframes, arreglos, conjuntos, tupla

```python
df5.sort_values(by =['Freq'],ascending=False).reset_index(drop=True).head(5)
```

```python
df5.Species.tolist()
```

```python
Counter(df5.Species.tolist())
```

```python
# Counter me regresa la frecuencia
orden = dict(Counter(df5.Species.tolist()))
```

```python
orden
```

```python
dat = {} # los diccionarios son estructuras de datos delimitadas por llaves
for v in list(reversed(sorted(set(orden.values())))):
    for x in orden:
        if orden[x] == v:
            print(v, x)
            dat[x] = v
```

```python
dat
```

```python
list(dat.keys())
```

```python
list(dat.values())
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(20, 5))

ax = fig.add_axes([0, 0, 1, 1])
ax.set_xlim(-1, len(dat))

ax.set_title("Análisis de abundancia", fontsize = 20)

ax.bar(list(dat.keys()), list(dat.values()), color ='red', width = 0.8)

ax.set_xlabel("Bacterias", fontsize = 20)
ax.set_ylabel("Frecuencia", fontsize = 20)

ax.set_xticklabels(list(dat.keys()), rotation = 90, ha= 'center', va = 'top', fontsize = 10, style = 'italic')
ax.yaxis.set_tick_params(labelsize=12, rotation = 0)

plt.show()
```

## Normalizar la escala con logaritmo y volver a graficar

```python
dat3 = {}
for i in dat2:
    a = np.log10(dat2[i])
    dat3[i] = a
```

```python
dat3
```

```python
fig = plt.figure(figsize=(20, 5))

ax = fig.add_axes([0, 0, 1, 1])
ax.set_xlim(-1, len(dat3))

ax.set_title("Análisis de abundancia", fontsize = 20)

ax.bar(list(dat3.keys()), list(dat3.values()), color ='blue', width = 0.8)

ax.set_xlabel("Bacterias", fontsize = 20)
ax.set_ylabel("Frecuencia", fontsize = 20)

ax.set_xticklabels(list(dat3.keys()), rotation = 90, ha= 'center', va = 'top', fontsize = 10, style = 'italic')
ax.yaxis.set_tick_params(labelsize=12, rotation = 0)

plt.show()
```

# Abundancia relativa
## volver a graficar aquellas con abundancia de más del 1%

```python
suma = np.sum(list(dat.values()))
```

```python
suma
```

```python
dat4 = {}
for i in dat:
    b = (dat[i] / suma) * 100
    dat4[i] = b
```

```python
dat5 = {}
for i in dat4:
    if dat4[i] >= 1:
        print(i, dat4[i])
        dat5[i] = dat4[i]
```

```python
dat5
```

### En el siguiente gráfico se muestran aquellas especies con una abundancia relativa mayor o igual al 1%

```python
fig = plt.figure(figsize=(7, 3))

ax = fig.add_axes([0, 0, 1, 1])
ax.set_xlim(-1, len(dat5))

ax.set_title("Análisis de abundancia", fontsize = 20)

ax.bar(list(dat5.keys()), list(dat5.values()), color ='limegreen', width = 0.8)

ax.set_xlabel("Bacterias", fontsize = 20)
ax.set_ylabel("Abundancia relativa", fontsize = 20)

ax.set_xticklabels(list(dat5.keys()), rotation = 90, ha= 'center', va = 'top', fontsize = 10, style = 'italic')
ax.yaxis.set_tick_params(labelsize=12, rotation = 0)

plt.show()
```
